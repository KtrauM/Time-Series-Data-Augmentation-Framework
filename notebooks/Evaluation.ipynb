```python
import os
import re
import numpy as np
```

```python
logfile_path = r"C:\Users\Murat\Desktop\Bachelor\saved_models\Run_logs\logs"'\\'
```

```python
experiments = os.listdir(logfile_path)
```

# average over subjects

```python
for e in experiments:
    experiment_path = logfile_path + e
    score_path = experiment_path + '/finetuned_score.txt'
    if not os.path.exists(score_path):
        score_path = experiment_path + '/score.txt'
    # print('score path')
    # print(experiment_path)
    # print(os.path.exists(experiment_path))
    if os.path.exists(score_path):
        print(f'Processing {e}...')
        
        weighted_F1 = []
        macro_F1 = []
        
        f = open(score_path)
        while True:
            line = f.readline()
            if line == '':
                break
            scores = re.findall(r'0.[\d.]*\d+', line)
            weighted_F1.append(float(scores[0]))
            macro_F1.append(float(scores[1]))
        
        F1 = np.array([weighted_F1, macro_F1]).T
        mean_F1 = np.mean(F1, axis=0)
        # print('zort')
        # print(experiment_path)
        # print(os.path.exists(experiment_path))
        # with open(os.path.join(experiment_path, 'avg.txt'), 'w') as fp:
        #     pass
        # for f in os.listdir(experiment_path):
        #     print(f)
        # print(os.path.join(experiment_path, 'average_score.txt'))
        np.savetxt(os.path.join(experiment_path, 'avg.txt'), mean_F1, fmt='%.6f')
        print('done once with path ' + os.path.join(experiment_path, 'avg.txt'))
        print('--------------')
```

# Average over seeds

```python
models = ['deepconvlstm']
datasets = ['hapt']
sfs = ['1']
```

```python
for model in models:
    for dataset in datasets:
        for sf in sfs:
            score_cross_seed = []
            for e in experiments:
                if (model in e) and (dataset in e) and (sf in e):
                    print(e)
                    seed_path = logfile_path + e + '/avg.txt'
                    score = np.loadtxt(seed_path)
                    print(score)
                    score_cross_seed.append(score)
            m = np.mean(score_cross_seed, axis=0)
            std = np.std(score_cross_seed, axis=0)
            result_cross_seed = np.array([m, std])
            np.savetxt(f'{logfile_path}/../result/{model}_{dataset}_{sf}.txt', result_cross_seed, fmt='%.6f')
```
